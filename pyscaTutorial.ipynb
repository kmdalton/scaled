Now we want to convert every residue in the alignment matrix to a corresponding number. To do so we arbitrarily assign a unique number (0-19) for all 20 amino acids. The number 20 represents a blank character "-" usedd to represent a gap in the alignent.

We import and visualize the alignment file using fullmsa.binMatrix, which has a fasta file importer (located in fasta.py)

```python
algnfile = '../sca5/Inputs/al_pdz.free '
# Note that this throws out sequences larger than the mode sequence length
aamtx = fullmsa.binMatrix(algnfile)

# Let's see what this looks like
f,ax1 = subplots()
ax1.imshow(aamtx.T,extent=[0,1,0,1])
ax1.set_xlabel('Sequences', size=12)
ax1.set_ylabel('Residue position', size=12)
ax1.set_title('Alignment of PDZ domain sequences', size=12)
```

From this alignment we can come up with a concensus sequence. 

```python
print "the PDZ family concensus sequence is \"" + fullmsa.consensus(aamtx) + "\""
```

As we cam see, there are a lot of gap characters in the concensus sequence, which suggests that the minority of sequences have inseretions at those positions. To fix this, we prune the matrix (and concensus sequence) of those gap characters.

```python
# Question: what's the difference in prune() vs newPrune()?

aamtx_pruned = fullmsa.prune(aamtx)

f,ax1 = subplots()
ax1.imshow(aamtx_pruned.T,extent=[0,1,0,1])
ax1.set_xlabel('Sequences', size=12)
ax1.set_ylabel('Residue position', size=12)
ax1.set_title('Pruned PDZ domain sequences', size=12)

print "the PDZ pruned family concensus sequence is \"" + fullmsa.consensus(aamtx_pruned) + "\""
```
