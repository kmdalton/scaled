Now we want to convert every residue in the alignment matrix to a corresponding number. To do so we arbitrarily assign a unique number (0-19) for all 20 amino acids. The number 20 represents a blank character "-" usedd to represent a gap in the alignent.

We import and visualize the alignment file using fullmsa.binMatrix, which has a fasta file importer (located in fasta.py)

```python
algnfile = 'al_pdz.free'
# Note that this throws out sequences larger than the mode sequence length
aamtx = fullmsa.binMatrix(algnfile)

# Let's see what this looks like
f,ax1 = subplots()
ax1.imshow(aamtx.T,extent=[0,1,0,1])
ax1.set_xlabel('Sequences', size=12)
ax1.set_ylabel('Residue position', size=12)
ax1.set_title('Alignment of PDZ domain sequences', size=12)
```

From this alignment we can come up with a concensus sequence. 

```python
print "the PDZ family concensus sequence is \"" + fullmsa.consensus(aamtx) + "\""
```

As we cam see, there are a lot of gap characters in the concensus sequence, which suggests that the minority of sequences have insertions at those positions. Since these regions are information-poor, we prune the matrix (and concensus sequence) of those gap characters.

```python
# Question: what's the difference in prune() vs newPrune()?

aamtx_pruned = fullmsa.prune(aamtx) # There's more than one way to skin a cat...

f,ax1 = subplots()
ax1.imshow(aamtx_pruned.T,extent=[0,1,0,1])
ax1.set_xlabel('Sequences', size=12)
ax1.set_ylabel('Residue position', size=12)
ax1.set_title('Pruned PDZ domain sequences', size=12)

concseq = fullmsa.consensus(aamtx_pruned)

print "the PDZ pruned family concensus sequence is \"" + concseq + "\""
```

Next we look at the positional conservation of the protein sequence. This is captured using the Kullback-Leibler relative entropy metric $D(f(a,i)||q(a))$, which measures the difference in frequency of amino acids at each residue relative to background frequencies. It will be the metric we use to attribute significance to each residue.

Here, one of the assumptions made by Ranganathan is that background frequency of residues can be estimated by looking over the non-repeating protein database. However, residue frequencies are known to vary widely for different species, and even specific proteins. Another possible estimate for background frequencies is from the actual protein of interest. Lets see how background frequency changes our measure of the Kullback-Leibler relative entropy.  We'll also compare the results when we use uniform background frequencies.


```python
# The hard-coded background residue frequencies estiamted from the non-repeating protein database.
BGQ = np.array([0.073, 0.025, 0.050, 0.061, 0.042, 0.072, 0.023, 0.053, 0.064, 0.089, 0.023, 0.043, 0.052, 0.040, 0.052, 0.073, 0.056, 0.063, 0.013, 0.033,1./21.])
BGQ_uniform = np.ones(BGQ.size)/21.

# The estimated background residue frequency for our protein of interest
BGQ2 = fullmsa.estimateBGFreq(aamtx_pruned)
f,(ax1,ax2,ax3) = subplots(3,1,figsize=[10,15])
ax1.plot(BGQ,label="non-repeating protein database")
ax1.plot(BGQ2,label="alignment matrix")
ax1.plot(BGQ_uniform,label="uniform frequency")
ax1.set_xlabel('Amino Acid', size=12)
ax1.set_ylabel('Frequency', size=12)
ax1.legend()

ax2.plot(fullmsa.DKL2(aamtx_pruned,BGQ),label="non-repeating protein database")
ax2.plot(fullmsa.DKL2(aamtx_pruned,BGQ2),label="alignment matrix")
ax2.plot(fullmsa.DKL2(aamtx_pruned,BGQ_uniform),label="uniform frequency")
ax2.set_xlabel('Residue Position', size=12)
ax2.set_ylabel('Kullback-Liebler Divergence', size=12)
ax2.set_xlim([0,aamtx_pruned.shape[1]])
ax2.legend()

ax3.hist(fullmsa.DKL2(aamtx_pruned,BGQ2))
ax3.set_xlabel('Kullback-Liebler Divergence', size=12)
ax3.set_ylabel('Frequency', size=12)
```

These small changes in entropy introduce uncertainty that will propogate through the rest of this analysis.

<h3> Finally, we can do some ICA</h3>
The departure of pysca from Rama's SCA implementation is in the manner that the pair-wise residue correlation matrix is calculated. Functionally, the main difference is instead of representing the amino-acid value as a position in a third binary dimension of size 21 (20 aa + 1 gap), we represent aa identity by an integer [0,20]. Hence, the matrix we have been visuallizing (aamtx_pruned) has column positions for each residue $\mathbf{M}_i$ , with each position  $M_{ij}\in{0,1,2...20}$ 

To construct the correlation matrix fr, we calculate a metric of the form <br>
$C_{ij} = \frac{H(\mathbf{M}_i)+H(\mathbf{M}_j)-2\cdot I(\mathbf{M}_i,\mathbf{M}_j)}{H(\mathbf{M}_i,\mathbf{M}_j)}$


where H is the Shannon entropy, and I is the mutual information. Note that this is not the same metric used as Rama. 

This metric is empirically chosen, and has the nice properties of satisfying the triangle inequality, non-negatiity and symmetry. The metric is also bounded between 0 and 1, with lower values representing more intimate relationships between two residue positions.

We also plot just the shannon entropy.

```python
if debugmode:
    reload(fullmsa)
    
infoDist = fullmsa.infoDistance(aamtx_pruned)
infoDist2 = fullmsa.testMetric(aamtx_pruned)

n = infoDist.shape[0]
infoDist[range(n), range(n)]= np.mean(np.mean(infoDist))


f,(ax1,ax2) = subplots(1,2,figsize=[10,5])
ax1.imshow(infoDist,extent=[0,1,0,1])
ax1.set_xlabel('Residue position', size=12)
ax1.set_ylabel('Residue position', size=12)
ax1.set_title('InfoDistance', size=12)
ax2.imshow(infoDist2,extent=[0,1,0,1])
ax2.set_xlabel('Sequences', size=12)
ax2.set_ylabel('Residue position', size=12)
ax2.set_title('InfoDistance', size=12)
```
