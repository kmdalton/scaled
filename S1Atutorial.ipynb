<h3>Step 3: Positional Correlation</h3>
We can measure the relative correlation of all residues using the Kullback-Leibler relative entropy. Ultimately, we will see how this value changes for all positions when each sequence is removed -- this will be the gradient of Kullback-Leibler entropy with respect to each sequence at each position.<br><br>
This measure will change depending on what we assume the background frequency is for each amino acid. In sca5, the non-repeating protein database is used to find this background frequency. However, we can also use the background frequency from our alignment matrix. How does this affect relative entropy?

```python
BGQ = np.array([0.073, 0.025, 0.050, 0.061, 0.042, 0.072, 0.023, 0.053, 0.064, 0.089, 0.023, 0.043, 0.052, 0.040, 0.052, 0.073, 0.056, 0.063, 0.013, 0.033,1./21.])
BGQ_uni = np.ones(BGQ.size)/21.

cons_hard = fullmsa.DKL2(aamtx_pruned,BGQ)
cons_uni = fullmsa.DKL2(aamtx_pruned,BGQ_uni)
cons_est = fullmsa.DKL(aamtx_pruned)


f,(ax1,ax2) = subplots(2,1,figsize=[10,10])
n,x = histogram(cons_hard,bins=20); x = (x[0:-1]+x[1:])/2.
n2 = np.array([float(n[i]) for i in range(n.size)])
ax1.plot(x,n2/np.sum(n2)/(x[1]-x[0]),label='sca5 background')
n,x = histogram(cons_uni,bins=20); x = (x[0:-1]+x[1:])/2.
n2 = np.array([float(n[i]) for i in range(n.size)])
ax1.plot(x,n2/np.sum(n2)/(x[1]-x[0]),label='uniform background')
n,x = histogram(cons_est,bins=20); x = (x[0:-1]+x[1:])/2.
n2 = np.array([float(n[i]) for i in range(n.size)])
ax1.plot(x,n2/np.sum(n2)/(x[1]-x[0]),label='protein background')
ax1.set_xlabel('Conservation', size=12)
ax1.set_ylabel('Frequency', size=12)
ax1.set_title('Pruned PDZ domain sequences', size=12)
ax1.legend()

ax2.plot(cons_hard, label='sca5 background')
ax2.plot(cons_uni, label='uniform background')
ax2.plot(cons_est, label='protein background')
ax2.set_xlabel('Residue Position')
ax2.set_ylabel('Conservation')
ax2.legend()
```

<h3>Step 4: SCA Calculation</h3>
This step finds a pairwise residue "statistical coupling energy" covariance matrix, where the metric defining the statistical coupling energy is the choice of the user. In the case of sca5, the metric is the so-called gradient of relative entropy, which measures the extent to which the amino acid at each residue from each sequence contributes to the relative entropy. This weight matrix (which is size $n_{seq} x n_{pos}$) is from what the position (Cp) or sequence (Cs) covariance matrix is calculated from.



```python
Cp,Cs,pwX = fullmsa.sca5(aamtx_pruned,BGQ)
f,((ax1,ax2),(ax3,ax4)) = subplots(2,2,figsize=[10,10])

ax1.imshow(pwX,extent=[0,1,0,1],interpolation='nearest')
ax1.set_xlabel('Position');ax1.set_ylabel('Sequence ID')
ax1.set_title('Weight Matrix')
ax2.imshow(Cp,extent=[0,1,0,1],interpolation='nearest')
ax2.set_xlabel('Position ID');ax2.set_ylabel('Position ID')
ax2.set_title('Position Covariance Matrix')
ax3.imshow(Cs,extent=[0,1,0,1],interpolation='nearest')
ax3.set_xlabel('Sequence ID');ax3.set_ylabel('Sequence ID')
ax3.set_title('Sequence Covariance Matrix')
ax4.imshow(M,extent=[0,1,0,1],interpolation='nearest')
ax4.set_xlabel('Sequence ID');ax4.set_ylabel('Sequence ID')
ax4.set_title('Sequence Similarity')
```

<h3>Step 5: Spectral Decomposition</h3>

To see how many significant modes exist in the residue covariance matrix, we compare the magnitude distribution of its eigenvalues to those of many randomized versions of the same matrix.


```python
import copy

wreal,vreal = np.linalg.eigh(Cp)

numiter = 20
wfake = np.zeros(wreal.size*numiter)

nseq,npos = pwX.shape

print "calculating random matrice eigen values " + str(numiter) + " times..."
for i in range(numiter):
    aln_rnd = np.transpose(copy.deepcopy(aamtx_pruned),[1,0])
    np.random.shuffle(aln_rnd)
    aln_rnd = np.transpose(aln_rnd,[1,0])
    Cptemp,Cstemp,pwXtemp = fullmsa.sca5(aln_rnd,BGQ)
    wtemp,vtemp = np.linalg.eigh(Cptemp)
    wfake[range(i*wtemp.size,(i+1)*wtemp.size)] = wtemp
    print i,
```

```python
f,(ax1) = subplots(1,figsize=[10,5])

n,x = histogram(wfake,bins=50); x = (x[0:-1]+x[1:])/2.
n2 = np.array([float(n[i]) for i in range(n.size)])
ax1.plot(x,n2/np.sum(n2)/(x[1]-x[0]),label='Random Distribution', linewidth=3)
n,x = histogram(wreal,bins=50); x = (x[0:-1]+x[1:])/2.
n2 = np.array([float(n[i]) for i in range(n.size)])
ax1.bar(x-(x[1]-x[0])/2,n2/np.sum(n2)/(x[1]-x[0]),width=(x[1]-x[0]),label='Real Distribution',color=[1, 0, 0, .5])
ax1.set_ylim([0,.1]); ax1.set_xlim([0,30])
ax1.set_ylabel('Frequency', fontsize=14)
ax1.set_xlabel('Magnitude of Eigenvalue', fontsize=14)
legend()
```

<h3>Step 5: Structure of the top Eigenmodes</h3>

Lets see how the top three eigenvectors look!

```python
w,v = np.linalg.eigh(Cp)

for i in range(w.size):
    v[:,i] *= np.sign(np.mean(v[:,i]))

idx = argsort(w)
idx = idx[::-1]

f,(ax1,ax2,ax3) = subplots(1,3,figsize=[15,5])
ax1.plot(v[:,idx[0]],v[:,idx[1]],linestyle='',marker='.')
ax1.set_xlabel('Eigen Vector 1'); ax1.set_ylabel('Eigen Vector 2')
ax2.plot(v[:,idx[0]],v[:,idx[2]],linestyle='',marker='.')
ax2.set_xlabel('Eigen Vector 1'); ax2.set_ylabel('Eigen Vector 3')
ax3.plot(v[:,idx[1]],v[:,idx[2]],linestyle='',marker='.')
ax3.set_xlabel('Eigen Vector 2'); ax3.set_ylabel('Eigen Vector 3')
```

In this case, the top mode (first eigenvector) demonstrates how conservation is related to the other two modes -- the larger the conservation, the larger the possible magnitude of the other modes. This is indicative of the weight that the sca5 matrix places on conservation in the weight matrix. 
<br><br>
On the rightmost graph, we see three possible distinct arms in the scatterplot between eigenvectors 2 and 3. To try and untangle these, we will use ICA.

<h3>Step 6: ICA</h3>

Lets decouple that last mode.

```python
from mdp.nodes import CuBICANode as ICANode

tokeep = 8

vec = np.real(np.asarray(v[:,idx[0:tokeep]]))
ica = ICANode(limit=0.00001)

ica.train(vec)

ic_P = ica(vec)

v1 = 0; v2 = 7; v3 = 5
f,(ax1,ax2,ax3) = subplots(1,3,figsize=[15,5])
ax1.plot(ic_P[:,v1],ic_P[:,v2],linestyle='',marker='.')
ax1.set_xlabel('Independent Component 1'); ax1.set_ylabel('Independent Component 2')
ax2.plot(ic_P[:,v1],ic_P[:,v3],linestyle='',marker='.')
ax2.set_xlabel('Independent Component 1'); ax2.set_ylabel('Independent Component 3')
ax3.plot(ic_P[:,v2],ic_P[:,v3],linestyle='',marker='.')
ax3.set_xlabel('Independent Component 2'); ax3.set_ylabel('Independent Component 3')
```
